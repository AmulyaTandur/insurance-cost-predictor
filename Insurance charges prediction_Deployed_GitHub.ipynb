Since the data has a normal distribustion , we use Feature scaling method to do Standardization

```python
from sklearn import preprocessing, model_selection 
# from sklearn.preprocessing import StandardScaler

labelencoder = preprocessing.LabelEncoder()   
scaler = preprocessing.StandardScaler()  
```

```python
health_df.isna().sum()        # Confirms , No NAN values 
```

```python
health_df.drop_duplicates(inplace=True)
```

```python
health_df = health_df[['age','sex','bmi','children','smoker','charges']]
```

### fit(), tranform() and fit_transform() -> vital tools that are extensively used in data preprocessing and model fitting.

```python
# Apply label encoding to SEX and Smoker variables 
from sklearn.preprocessing import LabelEncoder

le_sex = LabelEncoder()
le_smoker = LabelEncoder()

# Encode safely with .loc[] & Fit-transform and assign with explicit dtype conversion
health_df.loc[:,'sex'] = le_sex.fit_transform(health_df['sex']).astype('int64')
health_df.loc[:,'smoker'] = le_smoker.fit_transform(health_df['smoker']).astype('int64')
```

```python
health_df.head()
```

# Split X(independent variables) and Y(dependent variable)

```python

X = health_df.drop(['charges'], axis =1)
y = health_df[['charges']]
```

```python
X.columns
```

```python
scaled_X =scaler.fit_transform(X)     
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y,test_size =0.3, random_state =1)
```

```python
print("The shape of X_train is:",X_train.shape)
print("The shape of X_test is:",X_test.shape)
print("The shape of y_train is:",y_train.shape)
print("The shape of y_test is:",y_test.shape)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.formula.api as ols
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
print(model.score(X_train, y_train))

# This code in scikit-learn returns the coefficient of determination (R² score) by default for regression models.
```

### An R² score of 0.7534 means that only 75.3% of the variance in the target variable (charges) using the features in X_train.

```python
print(model.score(X_test, y_test))
```

###  Step 4: Make Predictions

```python
y_pred = model.predict(X_test)
```

### Step 5: Evaluate the Model

```python
from statsmodels.tools.eval_measures import rmse

R2 =r2_score(y_test,y_pred)
print('R2 Score between y_test and predicted :', R2)

RMSE = rmse(y_test,y_pred)
print('RMSE between y_test and predicted :', RMSE)
```

```python
%%writefile app.py       

#Now, you will have a file called "app.py" saved in the working directory.

import numpy as np
import pandas as pd
import streamlit as st
import pickle 

model=pickle.load(open('insurance_model.pkl','rb'))
scaler = pickle.load(open("scaler.pkl", "rb"))


def preprocess_input(age,sex, bmi, children, smoker):
    data = pd.DataFrame({"age": [age],
                         "sex": [sex],
                         "bmi": [bmi],
                         "children":[children],
                         "smoker": [smoker] })
    return data



def predict_insurance_charges(data):
    data_scaled = scaler.transform(data)     
    prediction = model.predict(data_scaled)
    return prediction

# Main app function
def main():
    st.title('Insurance Charges Prediction App')
    st.write("Enter customer details to predict insurance charges")
    st.sidebar.title('User Input')
    
    age = st.sidebar.slider("Age",20,100, step =1,value =30)    
                                                                
    sex = st.sidebar.selectbox("Sex",[0,1], format_func =lambda x: "Female" if x==0 else "Male")
    bmi = st.sidebar.slider("BMI", 10.0, 40.0, step =0.1, value =20.0)
    children = st.sidebar.slider('Number of Children', 0,10, step =1, value =0)
    smoker = st.sidebar.selectbox("Smoker",[0,1], format_func =lambda x: "No" if x==0 else "Yes")

    if st.button("Predict Charges"):
        input_data = preprocess_input(age,sex, bmi, children, smoker)
        prediction = predict_insurance_charges(input_data)
        
        st.subheader('Estimated Insurance Charge :')
        st.success(f"${float(prediction[0]):,.2f}")        

if __name__ == "__main__":
    main()
```
